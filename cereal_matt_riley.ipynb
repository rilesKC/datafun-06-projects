## Section 7-List: Get some of your information into a list. Process each item in the list (use for or comprehension as you like).

```python
low_calories = [calorie for calorie in df['calories'].tolist() if calorie <= 120]

print(f'List of low calories', (low_calories), end = '\n\n')

high_carbs = [carb for carb in df['carbo'].tolist() if carb > 14]

print(f'List of high carbs', (high_carbs), end = '\n\n')
```

## Section 8-Filter: Use filter() to show only part of the information.

## Section 9-Map: Use map() to transform some of the data.

```python
def convert_sugars_to_label(sugar):
    if sugar < 4:
        return 'Low'
    elif 4 <= sugar < 8:
        return 'Medium'
    else:
        return 'High'

df['sugar level'] = df['sugars'].map(convert_sugars_to_label)

print(df)
```
